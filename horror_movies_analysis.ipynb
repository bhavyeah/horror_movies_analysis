```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("horror_movies.csv")
```

```python
df.drop_duplicates(inplace = True)
df.dropna(inplace = True)
```

```python
print(f'we have : {df.duplicated().sum()} duplicated values')
```

```python
#changing the data type of release date 
df['release_date'] = pd.to_datetime(df['release_date'], dayfirst=True) #,  format='%B %d, %Y')
```

```python
#dropping unncessary columns
df.drop(['backdrop_path','poster_path','id'], axis=1)
```

```python
df['profit'] = df['revenue'] -  df['budget']
df['profit']
```

```python
hr = pd.DataFrame(df.groupby('original_title')[['revenue']].sum().sort_values('revenue', ascending=False)).head(10)
hr
```

```python
top_10_highest_profit = df[['title','profit']].sort_values(ascending = False, by = 'profit').head(10)
top_10_highest_profit
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_highest_profit['title'], y=top_10_highest_profit['profit'], palette="viridis") 
plt.xticks(rotation=45)  # better readability
plt.title('Top 10 Movies with the Highest Profits')
plt.xlabel('Movie Title')
plt.ylabel('Profit (in billions)')  
plt.tight_layout()  # subplot parameters for a better fit
plt.show()
```

```python
hp = pd.DataFrame(df.groupby('title')[['popularity']].sum().sort_values('popularity', ascending = False)).head(10)
hp
```
